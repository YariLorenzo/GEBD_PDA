```python
%matplotlib inline

import seaborn as sns
sns.set()

# Carga el ejemplo de datasets sobre vuelos
flights_long = sns.load_dataset("flights")
```

```python
flights_long.head()
```

```python
# pivotamos la tabla
flights = flights_long.pivot("month", "year", "passengers")
```

```python
flights.head()
```

```python
# Dibujamos un mapa de calor con los valores numéricos (pasajeros) en cada celda
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5)
```
